```julia
function read_tree(filename,c)
    lines = readlines(open(filename))
    count = Dict()
    for i = 1:length(lines)
        l = split(lines[i],c)
        count[l[1]] = 1 
        count[l[2]] = 1
    end
    
    n = length(count)
    G = SimpleGraph(n)
    for i = 1:length(lines)
        l = split(lines[i],c)
        add_edge!(G, parse(Int64,l[1])+1,parse(Int64,l[2])+1)
    end
    
    return G,count
end

function knn_acc(D,y)
    n,n = size(D)
    y_pred = zeros(n)
    for i = 1:n
        p = sortperm(D[i,:])
        if p[1] == i
            idx = p[2]
            y_pred[i] = y[idx]
        else
            idx = p[1]
            y_pred[i] = y[idx]
        end
    end
    
    return sum(y_pred .== y)/n
end   

function knn_acc_k(D,y,k)
    n,n = size(D)
    y_pred = zeros(n)
    for i = 1:n
        p = sortperm(D[i,:])
        counts = zeros(10)
        idx = 1
        kfound = 0
        while kfound < k
            if p[idx] == i
                idx+=1
            else
                counts[Int(y[p[idx]])+1] += 1
                kfound += 1
                idx += 1
            end
        y_pred[i] = argmax(counts)-1
        end
    end
    
    return sum(y_pred .== y)/n
end 

function remove_loops(G)
    n = nv(G)
    for i = 1:n
        if has_edge(G,i,i)
            rem_edge!(G,i,i)
        end
    end
    
    return G
end

function cal_distortion(D1,D2)
    a = Inf
    b = 0
    n,n = size(D2)
    for i = 1:n
        for j = 1:i-1
            if a > D1[i,j]/D2[i,j]
                a = D1[i,j]/D2[i,j]
            end
            if b < D1[i,j]/D2[i,j]
                b = D1[i,j]/D2[i,j]
            end
        end
    end
    
    return b/a
end


function avg_distortion(Dnew,Dold)
    n,n = size(Dnew)
    d = 0
    for i = 1:n
        for j = 1:i-1
            d += abs(Dnew[i,j]-Dold[i,j])/Dold[i,j]
        end
    end
    
    return 2*d/(n*(n-1))
end

function distance_with_missing(X,p=0.5)
    n,m = size(X)
    Q = rand(n,m) .> p
    D = zeros(n,n)
    for i = 1:n
        for j = 1:i-1
            D[i,j] = sum((Q[i,:].*Q[j,:].*(X[i,:]-X[j,:])).^2)
            D[j,i] = D[i,j]
        end
    end

  return D
end
```
